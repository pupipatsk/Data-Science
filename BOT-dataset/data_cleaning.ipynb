```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
```

```python
# BOT format
filepath = "data-raw/EC_EI_020_ENG_ALL.CSV"

df = pd.read_csv(filepath, index_col=0, encoding='utf-8')
df
```

```python
# columns datetime
# df.columns = df.columns.str[:-2]  # remove r(revised), p(preliminary)
# df.columns = pd.to_datetime(df.columns, format="%b %Y", errors="coerce")
df.columns = pd.to_datetime(df.columns, dayfirst=True, errors="coerce")
df.columns = df.columns + pd.offsets.MonthEnd()

# index features
df.index = df.index.str.strip()
df
```

```python
# Features Selection

# features = []
# df = df[df.index.isin(features)]

df = df.drop(['Exports', 'Imports'], axis=0)

df = df.transpose()
df.index.name = "date"

df
```

```python
# Save to Local
filename = 'Ex-Im Trade (USD)'

df.to_csv('data-prep/' + filename + '.csv',encoding='utf-8')
```
